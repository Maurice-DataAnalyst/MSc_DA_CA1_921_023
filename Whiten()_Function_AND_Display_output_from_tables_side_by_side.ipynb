```python
from IPython.display import display_html
from itertools import chain,cycle
```

### Create a method to display two dataframe tables side by side

```python
def display_side_by_side(*args,titles=cycle([''])):
    html_str=''
    
    for df,title in zip(args, chain(titles,cycle(['</br>'])) ):
    
        html_str+='<th style="text-align:center"><td style="vertical-align:top">'
        html_str+=f'<h3>{title}</h3>'
        html_str+=df.to_html().replace('table','table style="display:inline"')
        html_str+='</td></th>'
    
    display_html(html_str,raw=True)
```

```python
df_11 = pd.DataFrame(np.arange(12).reshape((3,4)),columns=['A','B','C','D',],)
df_22 = pd.DataFrame(np.arange(16).reshape((4,4)),columns=['V','W','X','Y',])
display_side_by_side(df_11,df_22,df_11, titles=['Chocolate','Toffee']) 
```

### Scale the data 
#### scipy.cluster.vq.whiten()

...........Each feature... (column) is divided by its standard deviation

across all observations... (rows) to give it......... unit variance

returns... an ndarray containing the values in obs scaled by the standard deviation of each column.


```python
# whiten the ndarray to_w and store in variable called w
w = vq.whiten(to_w)
```

```python
# what is the standard deviation of the 'First' column of df_2
df_2['First'].values.astype('float').std()
```

```python
# take the first value in the first row and cast it to a float and store it in a variable called some_value
some_value = (df_2['First'][:1]).values.astype('float')
```

```python
# store the standard deviation of the 'First' in a variable called col_std_dev
col_std_dev = df_2['First'].values.astype('float').std()
```

```python
# divide some_value by the standard deviation  
some_value/col_std_dev
```

##### Compare the output from the cell above to the first cell in std_dev

```python
whitened_df=pd.DataFrame(w,columns=['1st_std','2nd_std','3rd_std'])
```

#### Now add the 'Forth' column from the original dataframe df_2 to the standardized dataframe whitened_df

```python
whitened_df['Forth']=df_2['Forth']
```

#### Compare dataframe before and after whitened ---- using the value divided by the standard deviation of that column

```python
# note the magnitude of the column labeled '1st_std' with a tiny difference in the first value in the first row.
display_side_by_side(df_2,whitened_df, titles=['df_2','whitened'])
```
